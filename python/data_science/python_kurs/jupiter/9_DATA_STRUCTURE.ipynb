Слияние датафреймов
left        - объект DataFrame в левой части операции слияния
right       - объект DataFrame в правой части операции слияния
how         - допустимые значения  inner outer left right
on          - имена столбцов по которым производят соединение. Должны присутвовать в обоих объектах DataFrame
              если не заданы или не указаны никакие другие клчи соединения, 
              то используются имена столбцов общих для обоих объктов DataFrame
left_on     - столбцы левого DataFrame, используемы как ключи соединения
rigth_on    - столбцы правого DataFrame, используемы как ключи соединения
left_index  - использовать индекс левого DataFrame в качесве ключа соединения (или нескольких ключей в случае мультииндекса)
rigth_index - то же что и left_index только для праводго DataFrame
sort        - сортировать слитые данные лексикографически по ключам соединения по муолчанию True
              иногда в случае больших наборов лучше отключить 
suffixes    - кортеж строкб которые дописывают в конец совпадающих имен столбцов
copy        - если равен False то в некотрых осбых случаях разрешается не копировать даные в результирующую структуру

![типы соединений ](../../source/2.png)

```python
# у обоих датафреймов есть ключи и общая метка индекса 3
df1 = pd.DataFrame({'key': ['a', 'a', 'a', 'b', 'b', 'c', 'd'],
                    'data1': range(7)},
                    index = range(3,10))
df2 = pd.DataFrame({'key': ['a', 'b', 'b', 'f'],
                    'data2': range(4)})
```

```python
df1
```

```python
df2
```

```python
# по умолчанию merge использует inner
# соединение многие ко многоим по key 
# в результате попадают только те ключи которые присутвуют в обоих объектах
pd.merge(df1, df2)
```

```python
# по каким столбцам соединять 
# в примере выше не указано по каким столбцам соединять и поэтому используются столбцы с одинаковыми именами
# рекомендуется указывать столбец 
pd.merge(df1, df2, on='key')
```

```python
# разные имена столбцов
# если имена столбцов разные то задаются так
pd.merge(df1, df2, left_on='data1', right_on='data2')
```

```python
# задание суффиксов
# позволяют дописывать окончания к названиям столбцов
pd.merge(df1, df2, left_on='data1', right_on='data2', suffixes=('_left', '_right'))
```

```python
# соединение по нескольким столбцам
pd.merge(df1, df2,
         left_on=['data1', 'key'],
         right_on=['data2', 'key'])
```

```python
# указваем тип соединения
pd.merge(df1, df2, how='outer')
```

### Long Wide

```python
# long - формат используется в базах данных, экслель, csv
# Считываем данные акселерометра ("Long")
long_format = pd.read_csv("../../source/accel.csv")
long_format
```

```python
# в данном формате interval - промежуток времени 
# axis по которой производят замер ускорениая 
# reading - значение 
# проблема что очень тяжело выделить значения по конкреной оси 
# делают это так помещают interval, axis в индекс строк

# "Wide" фомат, используем unstack
tmp = long_format.set_index(['interval', 'axis'])
tmp
```

```python
# при использовании stack, unstack не происходит потери информации, только меняется способ организации
```

```python
#  unstack - поворачивает уровень строк преврая его в индекс столбцов
tmp = tmp.unstack()
tmp
```

```python
# еще один способ преобразовать данные из long_format --> wide_format
# "Wide" формат используем pivot
# интервалы - индексы строк
# столбцы - оси
# значниея берем из столбца reading   

wide_format = long_format.pivot(index='interval',
                                columns='axis',
                                values='reading')
wide_format
```

```python
# из Wide в long формат  обратное дейстие преращает уровень столбцов в уровень индексов строк
long_format_stack = wide_format.stack()
long_format_stack
```

```python
# создаем две копии данных акселерометра, по одной для каждого пользователя
user1 = long_format.copy()
user2 = long_format.copy()
```

```python
# добавляем столбец who в каждую копию
user1['who'] = 'Mike'
user2['who'] = 'David'
```

```python
# отмасштабируем данные user2
user2['reading']*=100
```

```python
user1.head()
```

```python
user2.head()
```

```python
# организуем данные так, чтобы получить иерархический индекс строк
# объединяем два датафрейма и задаем индексы
multi_user_long_format = pd.concat([user1, user2]).set_index(['who', 'interval', 'axis'])
multi_user_long_format
```

```python
# извлекаем все показания по всем осям и повсем пользователям при interva == 1
multi_user_long_format.xs(1, level='interval')
```

```python
# выполняем расстыковку, в результатае 2-й уровень индекса строк (уровень axis) стал уровнем индекса столбцов
multi_user_long_format = multi_user_long_format.unstack()
multi_user_long_format
```

```python
# параметр level для другого уровня 
multi_user_long_format.unstack(level=1)
```

```python
multi_user_long_format.unstack(level=0)
```
