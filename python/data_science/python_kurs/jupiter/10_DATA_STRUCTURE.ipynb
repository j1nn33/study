### Агрегирование групп

```python
# агрекирование - любое преобразование данных, которое порождает скаляное значение из массива
# вычиление среднего, кол-во значений
```

```python
tips.head(3)
```

```python
#произведем группировку ['day', 'smoker'] по  агрегированые данные в столюца tips_pct
group_tips = tips.groupby(['day', 'smoker'])['tips_pct']
group_tips
```

```python
# проверка эквивалентности методов
# для использованиая собственой функуии ее переадем в aggregate или кратко agg 
group_tips.agg == group_tips.aggregate
```

```python
# собственная функция (высислят разницу между мин и мах значением в массиве)
def peak_to_peak(arr):
    return arr.max() -arr.min()
group_tips.agg([peak_to_peak])
```

```python
# комбинирование 
group_tips.agg([peak_to_peak, 'mean'])
```

```python
# задаем имена столбцов
```

```python
group_tips.agg([('delta_max_min', peak_to_peak), ('mean_value', 'mean')])
```

```python
# отдельные наборы функция для каждого столбца # total_bill tips_pct
```

```python
group_tips = tips.groupby(['day','smoker'])['tips_pct', 'total_bill']
group_tips
```

```python
group_tips.agg({'tips_pct':[('max_value', np.max),
                            ('min_value', 'min')],
                'total_bill': 'sum'})
```

```python
# разультат имеет два уровня 
# 1 уровень tips_pct, total_bill
# 2 уровень max_value, min_value, sum
```

### Преобразование групп

```python
# метод transform
```

![transform ](../../source/4.png)

```python
df = pd.DataFrame({'Col1': ['A', 'B', 'C', 'C', 'B', 'B', 'A'],
                   'Col2': [1, 2, 3, 4, 2, 5, 3]})
df
```

```python
df['Col3'] = df.groupby('Col1').transform(sum)
df.sort_values('Col1')
```

```python
# объект groupby
```

```python
group_tips = tips.groupby(['smoker'])['total_bill']
group_tips
```

```python
# нормирование mean() - среднеиен std() - стандартное отколнение 
norm = lambda x: (x - x.mean())/x.std()
group_tips.transform(norm)
```

```python
# проверка 
tips_copy = tips.copy()
# добавляем новый столбец равный результату работы метода transform(norm)
tips_copy['total_bill_norm'] = group_tips.transform(norm)
tips_copy.head()
```

```python
tips_copy.groupby('smoker').total_bill_norm.agg(['mean', 'std'])
```

```python
# метод apply
# требуется выбрать первые 5 наиблольших значений tips_pct в каждой группе 
def top(df, n=5, column='tips_pct'):
    return df.sort_values(by=column)[-n:]
top(tips, n=6)
```

```python
tips.groupby('smoker').apply(top)
```

### Исключение групп

```python
df = pd.DataFrame({'Label': list('AABCCC'),
                   'Values': [1, 2, 3, 4, np.nan, 8]})
df
```

```python
# удаляем группы с менее чем двумя возможными значениями
# если true то влючетется в результат
```

```python
# удаляем группу с 1 заполненым сзанчением для нас это будет группа В
f = lambda x: x.Values.count() > 1
df.groupby('Label').filter(f)
```

```python
# удаляем группы, в которых есть пропуски
```

```python
f = lambda x: x.Values.isnull().sum() == 0
df.groupby('Label').filter(f)
```
