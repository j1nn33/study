##### Запуск потока с применением threading

внизу общее описание класса в виде объекта

```python
import threading
import time

# описание класса наш класс является наследником Thread
class MyThread(threading.Thread):
    def __init__(self, name, delay):
        threading.Thread.__init__(self)
        self.name = name
        self.delay = delay

    def run(self):
        print('Starting thread %s.' % self.name)
        thread_count_down(self.name, self.delay)
        print('Finished thread %s.' % self.name)

def thread_count_down(name, delay):
    counter = 3

    while counter:
        time.sleep(delay)
        print('Thread %s counting down: %i...' % (name, counter))
        counter -= 1
# Запуск одного потока имя, задежка 
#a=MyThread("Thread-1", 2,)
#a.start()
#print('Finished one tread.')

#### OUTPUT
#Starting thread Thread-1.
# Finished one tread.

#Thread Thread-1 counting down: 5...
#Thread Thread-1 counting down: 4...
#Thread Thread-1 counting down: 3...
#Thread Thread-1 counting down: 2...
#Thread Thread-1 counting down: 1...
#Finished thread Thread-1.
####################
# Запуск нескольких потков  имя, задежка 
a=MyThread("Thread-1", 2,)
# тк наш класс является наследником Thread, то запуск его производится через sart 
a.start()
b=MyThread("Thread-2", 3,)
b.start()
c=MyThread("Thread-3", 4,)
c.start()
```

Внизу алгоритм поиска простых чисел

```python
import threading
# Класс наследник от Thread поэтому запускается через start
class MyThread(threading.Thread):
    def __init__(self, x):
        threading.Thread.__init__(self)
        self.x = x

    def run(self):
        print('Starting processing %i...' % x)
        # вызов функции на проверку простое это число или нет
        is_prime(self.x)
        
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            print ('is not prime')
            return False
    return True 
```

```python
my_input = [2, 4, 193, 323, 1327, 433785907]

# список для хранения tread
threads = []

for x in my_input:
    temp_thread = MyThread(x)
    temp_thread.start()

    threads.append(temp_thread)
# после запуска всех tread
# процесс получения результата
for thread in threads:
    thread.join()

print('Finished.')
```

##### Передача данных между потоками при помощи pipe и общей памяти

Queue предоставляет нам механизм взаимодействия потоков между процессами FIFO (первым пришел — первым обслужен).


```python
from multiprocessing import Process, Queue
import random

def f(q):
    # функция делает только то, что закидывает в очередь какие-либо значения 42, None, 'hello'
    q.put([42, None, 'hello'])

def main():
    # Создаине очереди взаимодействивия между процессами
    q = Queue()
    # Создаем процесс который вычисляет функцию f и ему передается очередь q
    p = Process(target = f, args = (q,))
    p.start()
    # получение значений из очереди 
    print(q.get())
    
if __name__ == '__main__':
    main()
```

##### Pipe
Это структура данных, которая используется для связи между процессами в многопроцессорных программах.

Функция Pipe() возвращает пару объектов соединения, соединенных каналом, который по умолчанию является дуплексным (двухсторонним).

Работа Pipe происходит следующим образом. 

Pipe возвращает пару объектов соединения, которые представляют два конца канала.

У каждого объекта есть два метода — send() и recv() для взаимодействия между процессами.

```python
# то что получили выводим в обратном порядке 
from multiprocessing import Process, Pipe

def process2(conn):
    # отправляем в Pipe
    # данные будут доступны в другом процеесе
    conn.send([42, None, 'hello'])
    
    # получаем из Pipe данные в обратном порядке которые отправил y, если y не отправил ничего то процесс будет ожидать данные 
    x = conn.recv()
    print ('x reverse resive data from y ',x)
    conn.close()

if __name__ == '__main__':
    parent_conn, child_conn = Pipe()
    # создаем процесс который запускет функцию process2 и передаем ей child_conn
    # child_conn, - запятая тк это кортеж из одного элемента и она нужна
    p = Process(target = process2, args = (child_conn,))
    # создане процесса который выполянет функцию process2
    p.start()
    # получение данных которые получили от р при выполнении функции process2
    y = parent_conn.recv()
    print('y resive data from x', y)
    # отправка этих данных в обратном порядке
    parent_conn.send(y[::-1])
    # p.join()
```

##### Менеджер
Менеджер — это класс многопроцессорных модулей, который обеспечивает способ координации общей информации между всеми его пользователями. Управляющий объект управляет процессом сервера, который управляет общими объектами и позволяет другим процессам манипулировать ими. Другими словами, менеджеры предоставляют способ создавать данные, которые могут быть разделены между различными процессами. 

Ниже приведены различные свойства объекта менеджера —

Основным свойством менеджера является управление серверным процессом, который управляет общими объектами.

Другим важным свойством является обновление всех общих объектов, когда какой-либо процесс изменяет их.

Пример ниже использует объект менеджера для создания записи списка в процессе сервера, а затем добавления новой записи в этот список.

```python
import multiprocessing

def print_records(records):
    for record in records:
        print("Name: {0}\nScore: {1}\n".format(record[0], record[1]))

def insert_record(record, records):
    records.append(record)
    print("A New record is added\n")

if __name__ == '__main__':
    with multiprocessing.Manager() as manager:
        
        records = manager.list([('Computers', 1), ('Histoty', 5), ('Hindi',9)])
        new_record = ('English', 3)
        
        p1 = multiprocessing.Process(target = insert_record, args = (new_record, records))
        p2 = multiprocessing.Process(target = print_records, args = (records,))
        p1.start()
        p1.join()
        p2.start()
        p2.join()
```

Использование пространства имен для обмена данными между основным процессом и дочерним процессом.

```python
import multiprocessing

def Mng_NaSp(using_ns):
    
    using_ns.x +=5
    using_ns.y *= 10

if __name__ == '__main__':
    manager = multiprocessing.Manager()
    using_ns = manager.Namespace()
    using_ns.x = 1
    using_ns.y = 1
    
    print ('before', using_ns)
    p = multiprocessing.Process(target = Mng_NaSp, args = (using_ns,))
    p.start()
    p.join()
    print ('after', using_ns)
```

##### Ctypes-Array и Value
Многопроцессорный модуль предоставляет объекты Array и Value для хранения данных в карте общей памяти. Массив — это массив ctypes, выделенный из общей памяти, а Value — объект ctypes, выделенный из общей памяти.

```python
from multiprocessing import Process, Value, Array

def f(n, a):
    n.value = 3.1415927
    for i in range(len(a)):
        a[i] = -a[i]

if __name__ == '__main__':
    num = Value('d', 0.0)
    arr = Array('i', range(10))

    p = Process(target=f, args=(num, arr))
    p.start()
    p.join()

    print(num.value)
    print(arr[:])
```

##### Библиотека multiprocessing

```python
import os
from multiprocessing import Process
 
def doubler(number):
    """
    Функция умножитель на два
    """
    result = number * 2
    proc = os.getpid()
    print('{0} doubled to {1} by process id: {2}'.format(
        number, result, proc))
 
 
if __name__ == '__main__':
    numbers = [5, 10, 15, 20, 25]
    procs = []
    
    for index, number in enumerate(numbers):
        proc = Process(target=doubler, args=(number,))
        procs.append(proc)
        proc.start()
    
    for proc in procs:
        proc.join()
```

##### Логирование (Logging)

```python
import logging
import multiprocessing
from multiprocessing import Process, Lock

def printer(item, lock):
    """
    Выводим то что передали
    """
    lock.acquire()
    try:
        print(item)
    finally:
        lock.release()
 
 
if __name__ == '__main__':
    lock = Lock()
    items = ['tango', 'foxtrot', 10]
    multiprocessing.log_to_stderr()
    
    logger = multiprocessing.get_logger()
    logger.setLevel(logging.INFO)
    
    for item in items:
        p = Process(target=printer, args=(item, lock))
        p.start()
```
